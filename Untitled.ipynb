# Part 1: preparing the data

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
```

```python
#data is in same location as notebook
CATEGORIES=["bishop_resized","Rook-resize", "pawn_resized", "knight-resize","Queen-Resized"]
img_dir = "."

#iterate through examples in categories
for category in CATEGORIES:
    path = os.path.join(img_dir, category) #path to category directory
    for img in os.listdir(path):
        #convert to gray scale as data is smaller this way. Also adds nothing to this image classificationm model
        img_array=cv2.imread(os.path.join(path, img),cv2.IMREAD_GRAYSCALE)
        plt.imshow(img_array, cmap="gray") #load a sample image to see what it looks like
        plt.show()
        break
    break
        
```

```python
print(img_array.shape)
```

```python
print(img_array) #what our data looks like
```

```python
#normalise data by making eveyr image the same shape (in this case every image is 50 by 50)
IMG_SIZE = 50
new_array=cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array,cmap="gray" )
plt.show()
```

```python
#create the training dataset

training_data=[]

def create_training_data():

    ''' A function that creates the training data. Maps every unique chess peice name to a string'''

    for category in CATEGORIES:
        path = os.path.join(img_dir, category) #path to category directory
        class_num=CATEGORIES.index(category)
        
        for img in os.listdir(path):
                #convert to gray scale as data is smaller this way.
                # Also adds nothing to this image classificationm model 
                img_array=cv2.imread(os.path.join(path, img),cv2.IMREAD_GRAYSCALE)
                new_array =cv2.resize(img_array, (IMG_SIZE,IMG_SIZE))
                training_data.append([new_array, class_num])
            
```

```python
create_training_data()
```

```python
print(len(training_data))
```

```python
#must shuffle data so that model does not look at all pics of bishops, then all pics of knights, then all pics of pawns etc etc
import random

random.shuffle(training_data)
```

```python
print(training_data[0])
```

```python
#define variables that will be ved into the neural net:
X=[]
y=[]

for features, label in training_data:
    X.append(features)
    y.append(label)
#convert list to array to feed it into the NN

X=np.array(X).reshape(-1,IMG_SIZE, IMG_SIZE,1)
#how many features we have and the shape we have shaped our data in. 
# Also 1 at end because is gray scale
```

```python
import pickle # to save data

pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```

# Part 2: creating our model

```python
#import the needed libraries
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPool2D
```

```python
#Load the data in
X=pickle.load(open("X.pickle", "rb"))
Y=pickle.load(open("Y.pickle", "rb"))

# the next thing we need to do is scale the data. As the grayscale values
# range from 0 to 255, we can easily scale and normalise the data by dviding everything by 255:

X =X/255.0
```

```python
model = Sequential()

#window we want to use (3,3) zhape is not (-1,50,50,1) but (50,50,1) so shape[1:]
model.add(Conv2D(64, (3,3), input_shape= X.shape[1:]))

#after activation pass max pooling or activation
model.add(Activation("relu"))

#specify max pooling matrix
model.add(MaxPool2D(pool_size=(2,2)))

#flatten the matrix into a column vector to feed into regular nn
model.add(Flatten())

#feed column evctor into a 'regular' nn
model.add(Dense(64))
model.add(Dense(1)) #want one output

#can also specify this in the dense parameters
#multiclassification problem so softmax
model.add(Activation("softmax"))
```

```python
#configure parameters to compile and fit the model
callback = [tf.keras.callbacks.EarlyStopping(monitor='loss', patience=3),
             tf.keras.callbacks.ModelCheckpoint(filepath='best_model.h5', monitor='loss', save_best_only=True)]

#compile the model:
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
             loss = tf.keras.losses.CategoricalCrossentropy(), #categoricalcross entropy as multi classification problem
                metrics=["accuracy"])
```
